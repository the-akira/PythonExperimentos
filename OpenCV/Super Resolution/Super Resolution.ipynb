# Super Resolution com OpenCV

Aumentando a resolução de imagens sem perder a qualidade.

Primeiramente vamos importar as bibliotecas necessárias:

```python
import cv2
import matplotlib.pyplot as plt
from cv2 import dnn_superres
```

Criamos uma instância do objeto Super Resolution:

```python
sr = dnn_superres.DnnSuperResImpl_create()
```

Carregamos a apresentamos a imagem:

```python
image = cv2.imread("chinesepropaganda.png")
 
plt.figure(figsize=[12,12])
plt.imshow(image[:,:,::-1], interpolation='bicubic')
plt.axis('off');
```

Definimos o caminho do modelo que vamos utilizar a carregamos ele:

```python
model_path = "models/ESPCN_x2.pb"
sr.readModel(model_path)
sr.setModel("espcn", 2)
```

Aumentamos a resolução (upscale) da imagem:

```python
final_img = sr.upsample(image)
```
