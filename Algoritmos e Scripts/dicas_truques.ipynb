```python
names = ['Peter Parker', 'Clark Kent', 'Wade Wilson', 'Bruce Wayne']
heroes = ['Spiderman', 'Superman', 'Deadpool', 'Batman']

for index,name in enumerate(names):
    hero = heroes[index]
    print(f'{name} é {hero}')
```

```python
names = ['Peter Parker', 'Clark Kent', 'Wade Wilson', 'Bruce Wayne']
heroes = ['Spiderman', 'Superman', 'Deadpool', 'Batman']

for name,hero in zip(names, heroes):
    print(f'{name} é {hero}')
```

```python
names = ['Peter Parker', 'Clark Kent', 'Wade Wilson', 'Bruce Wayne']
heroes = ['Spiderman', 'Superman', 'Deadpool', 'Batman']
universes = ['Marvel', 'DC', 'Marvel', 'DC']

for name,hero,universe in zip(names, heroes, universes):
    print(f'{name} é {hero} do universo {universe}')
```

```python
# OOP

class Pessoa():
    pass

pessoa = Pessoa()

pessoa.primeiro = 'Gabriel'
pessoa.ultimo = 'Felippe'

print(pessoa.primeiro, pessoa.ultimo)
```

```python
class Pessoa():
    pass

pessoa = Pessoa()

chave = 'primeiro'
valor = 'Gabriel'

setattr(pessoa, 'primeiro', 'Gabriel')

print(pessoa.primeiro)
```

```python
class Pessoa():
    pass

pessoa = Pessoa()

chave = 'primeiro'
valor = 'Gabriel'

setattr(pessoa, chave, valor)

primeiro = getattr(pessoa, chave)

print(primeiro)
```

```python
class Pessoa():
    pass

pessoa = Pessoa()

pessoa_info = {'primeiro': 'Gabriel', 'segundo': 'Felippe'}

for chave, valor in pessoa_info.items():
    setattr(pessoa, chave, valor)

print(pessoa.primeiro, pessoa.segundo)
```

```python
class Pessoa():
    pass

pessoa = Pessoa()

pessoa_info = {'primeiro': 'Gabriel', 'segundo': 'Felippe'}

for chave, valor in pessoa_info.items():
    setattr(pessoa, chave, valor)

for chave in pessoa_info.keys():
    print(getattr(pessoa, chave))
```
